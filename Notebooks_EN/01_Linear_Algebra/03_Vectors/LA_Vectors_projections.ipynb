# Vectors - Projections and Divisions


## Projecting a Vector onto Another Vector

Projecting vector $\mathbf{a}$ onto axis $\mathbf{b}$ involves finding the component of vector $\mathbf{a}$ along the direction of $\mathbf{b}$.

Projection of $\mathbf{a}$ onto $\mathbf{b}$:

$$
\text{Proj}_{\mathbf{b}} \mathbf{a} = \frac{\mathbf{a} \cdot \mathbf{b}}{|\mathbf{b}|^2} \mathbf{b}
$$


## Projecting a Vector onto Axes

Projecting vector $\mathbf{a}$ onto an axis involves finding the component of vector $\mathbf{a}$ along that axis.

Projection of $\mathbf{a}$ onto the $OX$ axis:

$$
\text{Proj}_{OX} \mathbf{a} = \frac{\mathbf{a} \cdot \mathbf{i}}{|\mathbf{i}|^2} \mathbf{i} = \frac{\mathbf{a} \cdot \mathbf{i}}{1} \mathbf{i} = (\mathbf{a} \cdot \mathbf{i}) \mathbf{i}
$$

Projection of $\mathbf{a}$ onto the $OY$ axis:

$$
\text{Proj}_{OY} \mathbf{a} = \frac{\mathbf{a} \cdot \mathbf{j}}{|\mathbf{j}|^2} \mathbf{j} = \frac{\mathbf{a} \cdot \mathbf{j}}{1} \mathbf{j} = (\mathbf{a} \cdot \mathbf{j}) \mathbf{j}
$$

Recall that $\mathbf{i} = (1, 0)$ and $\mathbf{j} = (0, 1)$ are unit vectors along the $OX$ and $OY$ axes.


## Dividing a Line Segment in a Given Ratio

A line segment is defined by two points in space.

Line segments and vectors are closely related because a line segment connecting two points can be treated as a vector directed from one point to another.

If point \( P \) divides the line segment connecting points $A(x_1, y_1)$ and $B(x_2, y_2)$ in the ratio $m:n$, then the coordinates of point $P$ can be found using the formula:

$$
P = \left(\frac{n \times x_2 + m \times x_1}{m+n}, \frac{n \times y_2 + m \times y_1}{m+n}\right)
$$

**Example:**

Given a line segment connecting points $A(1, 2)$ and $B(3, 8)$, find point $P$ that divides the segment in the ratio 2:3.

Coordinates of point $P$:
$$
P = \left(\frac{2 \times 3 + 3 \times 1}{2+3}, \frac{2 \times 8 + 3 \times 2}{2+3}\right) = \left(\frac{9}{5}, \frac{22}{5}\right)
$$

```python
import matplotlib.pyplot as plt
import math

# Coordinates of points A and B
A = (1, 2)
B = (3, 8)

# Ratio m:n
m, n = 2, 3

# Calculate coordinates of point P that divides the segment in the ratio 2:3
P_x = (n * A[0] + m * B[0]) / (m + n)
P_y = (n * A[1] + m * B[1]) / (m + n)
P = (P_x, P_y)

# Plotting the graph
plt.figure(figsize=(6, 6))

# Plotting line segment AB
plt.plot([A[0], B[0]], [A[1], B[1]], 'k--', label='Line Segment AB')

# Marking points A, B, and P
plt.scatter(*A, color='blue', label='A(1, 2)')
plt.scatter(*B, color='red', label='B(3, 8)')
plt.scatter(*P, color='green', label=f'P({P[0]:.2f}, {P[1]:.2f})')

# Point labels
plt.text(A[0]-0.2, A[1], 'A(1, 2)', fontsize=12, color='blue')
plt.text(B[0]+0.1, B[1], 'B(3, 8)', fontsize=12, color='red')
plt.text(P[0]+0.1, P[1], f'P({P[0]:.2f}, {P[1]:.2f})', fontsize=12, color='green')

# Adding grid and axis labels
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Dividing a Line Segment in the Ratio 2:3')
plt.legend()

# Displaying the graph
plt.show()
```

```python
# Define vector a
a = (3, 4)
x = (1, 0)
y = (0, 1)
# Projection onto OX axis
proj_x = (a[0] * x[0], 0 * x[1])  # The x-component remains the same, y-component becomes 0
# the result multiplied by x stays the same

# Projection onto OY axis
proj_y = (a[0] * y[0], a[1]* y[1])  # The x-component becomes 0, y-component remains the same
# the result multiplied by y stays the same

print("OX axis projection:", proj_x)
print("OY axis projection:", proj_y)
```

```python
# Plotting the graph
plt.figure(figsize=(6, 6))

# Plot vector a
plt.quiver(0, 0, 3, 4, angles='xy', scale_units='xy', scale=1, color='blue', label='Vector a')

# Plot projection onto OX axis
plt.quiver(0, 0, 3, 0, angles='xy', scale_units='xy', scale=1, color='red', label='Projection onto OX')

# Plot projection onto OY axis
plt.quiver(0, 0, 0, 4, angles='xy', scale_units='xy', scale=1, color='green', label='Projection onto OY')


# Adding grid, labels, title and legend
plt.grid(True)
plt.xlim(-1, 5)
plt.ylim(-1, 6)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Projection of Vector a onto OX and OY axes')
plt.legend()

# Display the plot
plt.show()
```

```python
import sympy as sp
# Define vectors a and b
a = (2, 3)
b = (1, 1)

# Calculate the dot product of a and b
dot_product = a[0] * b[0] + a[1] * b[1]

# Calculate the squared magnitude of b
magnitude_b_squared = sp.sqrt(b[0]**2 + b[1]**2)**2

# Calculate the projection of a onto b
proj_a_b = (dot_product / magnitude_b_squared) # a.b/||b||**2
proj_a_x = (proj_a_b * b[0], proj_a_b * b[1]) # result . b

# Convert real number to float
proj_a_x = (float(proj_a_x[0]), float(proj_a_x[1]))

print("Projection of a onto b:", proj_a_x)
```

```python
# Plotting the graph
plt.figure(figsize=(6, 6))

# Plot vector a
plt.quiver(0, 0, 2, 3, angles='xy', scale_units='xy', scale=1, color='blue', label='Vector a')

# Plot vector b
plt.quiver(0, 0, 1, 1, angles='xy', scale_units='xy', scale=1, color='red', label='Vector b')

# Plot projection of a onto b
plt.quiver(0, 0, proj_a_x[0], proj_a_x[1], angles='xy', scale_units='xy', scale=1, color='green', label='Projection of a onto b')

# Adding grid, labels, title and legend
plt.grid(True)
plt.xlim(-1, 3)
plt.ylim(-1, 4)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Projection of Vector a onto Vector b')
plt.legend()

# Display the plot
plt.show()
```

```python
import sympy as sp
# Define vectors b and a
b = (1, 1)
a = (2, 3)

# Calculate the dot product of b and a
dot_product = b[0] * a[0] + b[1] * a[1]

# Calculate the squared magnitude of a
magnitude_a_squared = sp.sqrt(a[0]**2 + a[1]**2)**2

# Calculate the projection of b onto a
proj_b_a = (dot_product / magnitude_a_squared.evalf()) # b.a/||a||**2
proj_b_a = (proj_b_a * a[0], (proj_b_a * a[1])) # result . a

print("Projection of a onto b:", proj_b_a)
```

```python
# Plotting the graph
plt.figure(figsize=(6, 6))

# Plot vector b
plt.quiver(0, 0, 1, 1, angles='xy', scale_units='xy', scale=1, color='red', label='Vector b')

# Plot vector a
plt.quiver(0, 0, 2, 3, angles='xy', scale_units='xy', scale=1, color='blue', label='Vector a')

# Plot projection of b onto a
plt.quiver(0, 0, proj_b_a[0], proj_b_a[1], angles='xy', scale_units='xy', scale=1, color='green', label='Projection of b onto a')

# Adding grid, labels, title and legend
plt.grid(True)
plt.xlim(-1, 3)
plt.ylim(-1, 4)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Projection of Vector b onto Vector a')
plt.legend()

# Display the plot
plt.show()
```
