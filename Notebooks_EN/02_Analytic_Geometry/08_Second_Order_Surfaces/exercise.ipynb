### **1. Write the equation of a sphere with center at point $$P = (1, 2, 3)$$ and radius $$r = 3$$:**

**Question:**  
Write the equation of a sphere with center at point $$P = (1, 2, 3)$$ and radius $$r = 3$$.

**Explanation:**  
The general equation of a sphere with center at $$(x_0, y_0, z_0)$$ and radius $$r$$ is:

$$
(x - x_0)^2 + (y - y_0)^2 + (z - z_0)^2 = r^2
$$

Substituting the values for the center $$(x_0, y_0, z_0) = (1, 2, 3)$$ and radius $$r = 3$$, we get:

$$
(x - 1)^2 + (y - 2)^2 + (z - 3)^2 = 9
$$

**Question:**

The equation $$x^2 + y^2 + z^2 = 2$$ represents a sphere with center at the origin and radius $$\sqrt{2}$$.

Since the second sphere has a larger radius ($$\sqrt{2} > 1$$), the spheres will intersect, and they share common points. To find these points, we need to visualize the intersection.

The equations are:

$$
x^2 + y^2 + z^2 = 1
$$
$$
x^2 + y^2 + z^2 = 2

### **3. What curve in space is formed by the intersection of the sphere $$x^2 + y^2 + z^2 = 1$$ with the sphere $$(x - 1)^2 + y^2 + z^2 = 1$$? Find the equation of this curve.**

**Question:**  
What curve in space is formed by the intersection of the sphere $$x^2 + y^2 + z^2 = 1$$ with the sphere $$(x - 1)^2 + y^2 + z^2 = 1$$? Find the equation of this curve.

**Explanation:**  
The first sphere is centered at the origin with radius 1:

$$
x^2 + y^2 + z^2 = 1
$$

The second sphere is centered at $$(1, 0, 0)$$ with radius 1:

$$
(x - 1)^2 + y^2 + z^2 = 1
$$

To find the curve of intersection, subtract the second equation from the first:

$$
(x^2 + y^2 + z^2) - ((x - 1)^2 + y^2 + z^2) = 0
$$

Simplifying the equation gives the equation of the curve formed by the intersection of the two spheres.

### **4. Write the equation of the tangent plane to the paraboloid $$z = (x - 1)^2 + y^2 + 1$$ at point $$P = (1, 0, 1)$$:**

**Question:**  
Write the equation of the tangent plane to the paraboloid $$z = (x - 1)^2 + y^2 + 1$$ at point $$P = (1, 0, 1)$$.

**Explanation:**  
The equation of the paraboloid is given by:

$$
z = (x - 1)^2 + y^2 + 1
$$

To find the tangent plane at the point $$P = (1, 0, 1)$$, we first compute the partial derivatives $$f_x$$ and $$f_y$$ of the function $$z = (x - 1)^2 + y^2 + 1$$. These derivatives give the slopes of the tangent plane in the $$x$$ and $$y$$ directions, respectively.

The general equation for the tangent plane at point $$P(x_0, y_0, z_0)$$ is:

$$
z - z_0 = f_x(x_0, y_0)(x - x_0) + f_y(x_0, y_0)(y - y_0)
$$

Substituting $$P = (1, 0, 1)$$ and the partial derivatives of $$z = (x - 1)^2 + y^2 + 1$$, we get the equation of the tangent plane.
