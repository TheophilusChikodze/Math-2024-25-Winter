# Question 7
Find the distance from point $$ A(1, 2, 3) $$ to the plane $$ 2x + 3y + 4z = 5 $$.

The formula for the distance from a point $$ (x_1, y_1, z_1) $$ to a plane $$ Ax + By + Cz + D = 0 $$ is:

$$
d = \frac{|Ax_1 + By_1 + Cz_1 + D|}{\sqrt{A^2 + B^2 + C^2}}
$$

For the plane $$ 2x + 3y + 4z = 5 $$, we have $$ A = 2, B = 3, C = 4, D = -5 $$ and point $$ A(1, 2, 3) $$.

Substitute into the formula:

$$
d = \frac{|2(1) + 3(2) + 4(3) - 5|}{\sqrt{2^2 + 3^2 + 4^2}} = \frac{|2 + 6 + 12 - 5|}{\sqrt{4 + 9 + 16}} = \frac{|15|}{\sqrt{29}} = \frac{15}{\sqrt{29}}
$$

# Question 9
Calculate the angle between the plane $x + y + z = 1$ and the plane $x = 0$ (the $yz$-plane)

\\

The angle between two planes is determined by the angle between their normal vectors. Let the normal vectors of the two planes be:

1. For the plane $$x + y + z = 1$$, the normal vector is:  
   $$\vec{n_1} = (1, 1, 1)$$

2. For the plane $x = 0$
(the $yz$-plane), the normal vector is:  
   $$\vec{n_2} = (1, 0, 0)$$

The angle $$\theta$$ between the two planes can be calculated using:  
$$\cos\theta = \frac{|\vec{n_1} \cdot \vec{n_2}|}{||\vec{n_1}|| \cdot ||\vec{n_2}||}$$

### Solution:
1. Compute the dot product:  
   $$\vec{n_1} \cdot \vec{n_2} = (1)(1) + (1)(0) + (1)(0) = 1$$

2. Compute the magnitudes of the normal vectors:  
   $$||\vec{n_1}|| = \sqrt{1^2 + 1^2 + 1^2} = \sqrt{3}$$  
   $$||\vec{n_2}|| = \sqrt{1^2 + 0^2 + 0^2} = 1$$

3. Substitute into the formula for $$\cos\theta$$  
   $$\cos\theta = \frac{|1|}{\sqrt{3} \cdot 1} = \frac{1}{\sqrt{3}}$$

4. Calculate
   $$\theta = \arccos\left(\frac{1}{\sqrt{3}}\right)$$


The angle between the two planes is:  54.74$^\circ$
