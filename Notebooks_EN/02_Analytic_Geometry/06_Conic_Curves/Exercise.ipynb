# 1. Equation of a Circle
**Question:** Find the equation of the circle with center \( A(1, 2) \) and radius \( r = 3 \).

The general equation of a circle is:
$$
(x - h)^2 + (y - k)^2 = r^2
$$
For center \( A(1, 2) \) and radius \( r = 3 \):
$$
(x - 1)^2 + (y - 2)^2 = 9
$$

# 2. Equation of a Parabola
**Question:** Find the equation of the parabola that intersects the \( O_x \)-axis at \( x = 2 \) and \( x = 4 \), and passes through the point \( (3, 1) \).

The equation of a parabola intersecting the \( O_x \)-axis at \( x = 2 \) and \( x = 4 \) is:
$$
y = a(x - 2)(x - 4)
$$
Substitute \( (3, 1) \) to find \( a \):
$$
1 = a(3 - 2)(3 - 4)
$$
$$
1 = a(-1) \implies a = -1
$$
The equation is:
$$
y = -(x - 2)(x - 4)
$$

# 3. Center of the Ellipse
**Question:** Find the center of the ellipse represented by the equation \( x^2 + 4y^2 - 4x - 16y + 16 = 0 \).

Given equation:
$$
x^2 + 4y^2 - 4x - 16y + 16 = 0
$$
Group terms:
$$
(x^2 - 4x) + 4(y^2 - 4y) = -16
$$
Complete the square:
$$
(x - 2)^2 - 4 + 4((y - 2)^2 - 4) = -16
$$
Simplify:
$$
(x - 2)^2 + 4(y - 2)^2 = 16
$$
The center is:
$$
(2, 2)
$$

# 4. Tangent Line to Circle
**Question:** Find the slope of the tangent line to the circle \( x^2 + y^2 = 1 \) given the line equation \( y = mx - 5 \).

Given circle:
$$
x^2 + y^2 = 1
$$
Line equation:
$$
y = mx - 5
$$
Substitute into the circle equation:
$$
x^2 + (mx - 5)^2 = 1
$$
Expand:
$$
x^2 + m^2x^2 - 10mx + 25 = 1
$$
Simplify:
$$
(1 + m^2)x^2 - 10mx + 24 = 0
$$
Tangency condition \( \Delta = 0 \):
$$
(-10m)^2 - 4(1 + m^2)(24) = 0
$$
Solve for \( m \):
$$
100m^2 - 96 - 96m^2 = 0 \implies 4m^2 = 96 \implies m^2 = 24 \implies m = \sqrt{24} = 2\sqrt{6}
$$

# 5. Intersection of Hyperbola and Ellipse
**Question:** Find the intersection points of the hyperbola \( x^2 - y^2 = 1 \) and the ellipse \( x^2 + 4y^2 = 6 \).

Hyperbola:
$$
x^2 - y^2 = 1
$$
Ellipse:
$$
x^2 + 4y^2 = 6
$$
Subtract:
$$
(x^2 + 4y^2) - (x^2 - y^2) = 6 - 1
$$
$$
5y^2 = 5 \implies y^2 = 1 \implies y = \pm 1
$$
Substitute \( y = \pm 1 \) into the hyperbola:
$$
x^2 - 1 = 1 \implies x^2 = 2 \implies x = \pm \sqrt{2}
$$
Intersection points:
$$
(\sqrt{2}, 1), (-\sqrt{2}, 1), (\sqrt{2}, -1), (-\sqrt{2}, -1)
$$

# 6. Distance Between Hyperbola Branches
**Question:** Find the distance between the branches of the hyperbola \( x^2 - y^2 = 1 \).

Hyperbola:
$$
x^2 - y^2 = 1
$$
General form:
$$
\frac{x^2}{a^2} - \frac{y^2}{b^2} = 1
$$
For \( a^2 = 1, b^2 = 1 \):
$$
b = 1
$$
Distance between branches:
$$
2b = 2(1) = 2
$$
