## Task 1: The line passes through points A(1, 2) and B(3, 4). Find the equation of the line.

To find the equation of a line passing through two points \( A(1, 2) \) and \( B(3, 4) \), we first calculate the slope \( m \) using the formula:
$$
m = \frac{y_2 - y_1}{x_2 - x_1}
$$
Then, we use the point-slope form of the line equation:
$$
y - y_1 = m(x - x_1)
$$

### Solution:
The slope \( m \) is:
$$
m = \frac{4 - 2}{3 - 1} = 1
$$
Using point \( A(1, 2) \), we substitute into the equation:
$$
y - 2 = 1(x - 1)
$$
Simplifying:
$$
y = x + 1
$$

Alternate Standard Form:
$$-x + y = 1
$$

## Task 2: The line passes through point A(1, 2) and is parallel to the line \( y = 2x + 3 \). Find the equation of the line.

Since the lines are parallel, they have the same slope. The slope of the line \( y = 2x + 3 \) is 2.

The equation of the line passing through \( A(1, 2) \) is:
$$
y - 2 = 2(x - 1)
$$
Simplifying:
$$
y = 2x
$$

## Task 3: The line passes through point A(1, 2) and is perpendicular to the line \( y = 2x + 3 \). Find the equation of the line.

The slope of a line perpendicular to \( y = 2x + 3 \) is the negative reciprocal of the slope 2. So, the perpendicular slope is \( m = -\frac{1}{2} \).

The equation of the line passing through \( A(1, 2) \) is:
$$
y - 2 = -\frac{1}{2}(x - 1)
$$
Simplifying:
$$
y = -\frac{1}{2}x + \frac{5}{2}
$$

## Task 4: We have two lines \( y = 2x + 3 \) and \( y = 3x + 2 \). Find the intersection point of these lines and calculate the angle between them.

To find the intersection, set the two equations equal to each other:
$$
2x + 3 = 3x + 2
$$
Solving for \( x \):
$$
x = 1
$$
Substitute \( x = 1 \) into one of the original equations to find \( y \):
$$
y = 2(1) + 3 = 5
$$
The intersection point is \( (1, 5) \).


To find the angle $$ \theta $$between the lines, we use the formula for the angle between two lines with slopes $$ m_1 $$ and $$ m_2 $$:
$$
\tan(\theta) = \left| \frac{m_1 - m_2}{1 + m_1 m_2} \right|
$$
For \( m_1 = 2 \) and \( m_2 = 3 \):
$$
\tan(\theta) = \left| \frac{2 - 3}{1 + 2 \times 3} \right| = \frac{1}{7}
$$
Thus, the angle is:
$$
\theta = \tan^{-1} \left( \frac{1}{7} \right)
$$

## Task 7: We have the line \( y = 2x + 3 \) and point A(1, 2). Find the distance from point A to the line.

The formula for the distance from a point \( (x_1, y_1) \) to a line \( Ax + By + C = 0 \) is:
$$
d = \frac{|Ax_1 + By_1 + C|}{\sqrt{A^2 + B^2}}
$$
Rewriting the line equation \( y = 2x + 3 \) to standard form:
$$
2x - y + 3 = 0
$$
For \( A = 2 \), \( B = -1 \), and \( C = 3 \), the distance is:
$$
d = \frac{|2(1) - 1(2) + 3|}{\sqrt{2^2 + (-1)^2}} = \frac{|3|}{\sqrt{5}} = \frac{3}{\sqrt{5}}
$$

## Task 8: The line intersects the coordinate axes at points A(2, 0) and B(0, 3). Find the equation of the line.

The equation of the line that intersects the x-axis at \( A(2, 0) \) and the y-axis at \( B(0, 3) \) is:
$$

## Task 9: Calculate the angle between the line \( y = x + 3 \) and the x-axis.

The slope of the line \( y = x + 3 \) is \( m = 1 \). The angle \( \theta \) between the line and the x-axis is given by:
$$
\tan(\theta) = m = 1
$$
Thus:
$$
\theta = \tan^{-1}(1) = 45^\circ
$$
