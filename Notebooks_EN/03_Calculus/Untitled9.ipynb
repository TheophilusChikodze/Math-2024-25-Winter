#Limits of Real Functions
##Question 1
###Compute: $$\lim_{x \to \infty} \frac{x^3 + 2x^2}{x^4 - 3x^3}$$

#### Solution:
To compute the limit:
$$
\lim_{x \to \infty} \frac{x^3 + 2x^2}{x^4 - 3x^3},
$$

\\

we divide the numerator and denominator by $x^4$, the highest power of $x$ in the denominator:

\\

$$
\frac{x^3 + 2x^2}{x^4 - 3x^3} = \frac{\frac{x^3}{x^4} + \frac{2x^2}{x^4}}{1 - \frac{3x^3}{x^4}}.
$$


Simplifying:
$$
\frac{x^3 + 2x^2}{x^4 - 3x^3} = \frac{\frac{1}{x} + \frac{2}{x^2}}{1 - \frac{3}{x}}.
$$

As $x \to \infty$, all terms with $\frac{1}{x}$ or $\frac{1}{x^2}$ approach $0$:
$$
\lim_{x \to \infty} \frac{x^3 + 2x^2}{x^4 - 3x^3} = \frac{0 + 0}{1 - 0} = 0.
$$

**Final Answer:**
$$
\boxed{0}
$$

##Question 2
### Find $$\lim_{x \to 0} \frac{\sin(3x)}{2x + 1}$$

#### Solution:
The limit is:
$$
\lim_{x \to 0} \frac{\sin(3x)}{2x + 1}.
$$

Using the small-angle approximation $$\sin(kx) \approx kx$$ as $x \to 0$ we rewrite the numerator:
$$
\lim_{x \to 0} \frac{\sin(3x)}{2x + 1} \approx \lim_{x \to 0} \frac{3x}{2x + 1}.
$$

\\

At $x = 0$, the function evaluates directly:
$$
\frac{3(0)}{2(0) + 1} = 0.
$$
Thus $$
\boxed{0}
$$

## 1. Asymptotes of $$f(x) = \frac{x^2 - 1}{x^2 + 1}$$

To determine the asymptotes:


- **Vertical asymptotes**: $f(x)$ is undefined when $x^2 + 1 = 0$. Since $x^2 + 1 > 0$ for all real $x$, there are no vertical asymptotes.


- **Horizontal asymptotes**: Compute: $$\lim_{x \to \infty} f(x)$$
$$
f(x) = \frac{x^2 - 1}{x^2 + 1} = \frac{1 - \frac{1}{x^2}}{1 + \frac{1}{x^2}}.
$$
As $$x \to \infty, \frac{1}{x^2} \to 0$$, so:
$$
\lim_{x \to \infty} f(x) = \frac{1 - 0}{1 + 0} = 1.
$$

\\

$y = 1$ is a horizontal asymptote.


**Thus,**
the function $f(x)$ has a horizontal asymptote at $y = 1$ and no vertical asymptotes.

### 2. Asymptotes of $$g(x) = \frac{\sin(x)}{x^2 + 1}$$


To determine the asymptotes:
- **Vertical asymptotes**: $g(x)$ is undefined when $x^2 + 1 = 0$. Since $x^2 + 1 > 0$ for all real $x$, there are no vertical asymptotes.


- **Horizontal asymptotes**: Compute; $$\lim_{x \to \infty} g(x)$$

\\

$$
g(x) = \frac{\sin(x)}{x^2 + 1}.
$$

\\

Since $\sin(x)$ oscillates between $-1$ and $1$, and $x^2 + 1 \to \infty$, the fraction approaches $0$:

\\

$$
\lim_{x \to \infty} g(x) = 0.
$$


Similarly:
$$
\lim_{x \to -\infty} g(x) = 0.
$$
$y = 0$ is a horizontal asymptote.


**Thus**, the function $g(x)$ has a horizontal asymptote at $y = 0$ and no vertical asymptotes.
