##Question 1
### Computation of the derivatives:  

1. $$y(x) = -3x + 3$$  
   **Derivative:**  
   $$y'(x) = -3$$  

2. $$y(x) = \pi x + \sin(1)$$  
   **Derivative:**  
   $$y'(x) = \pi$$ (since $\sin(1)$ is a constant)  

3. $$y(x) = 4 + \sin(2)$$  
   **Derivative:**  
   $$y'(x) = 0$$ (both terms are constants)  

4. $$y(x) = 2x^3 - 3x^2 + 8x - 9$$  
   **Derivative:**  
   $$y'(x) = 6x^2 - 6x + 8$$  

5. $$y(x) = 6x^{1/3}$$  
   **Derivative:**  
   $$y'(x) = 2x^{-2/3} = \frac{2}{x^{2/3}}$$  

6. $$y(x) = x$$  
   **Derivative:**  
   $$y'(x) = 1$$  

7. $$y(x) = \cos(x) + \sin(x)$$  
   **Derivative:**  
   $$y'(x) = -\sin(x) + \cos(x)$$  

8. $$y(x) = 2\sin(x)\cos(x)$$  
   Using the double angle formula, $$y(x) = \sin(2x)$$  
   **Derivative:**  
   $$y'(x) = 2\cos(2x)$$  

9. $$y(x) = x\sin(x)$$  
   Using the product rule:  
   $$y'(x) = \sin(x) + x\cos(x)$$  

10. $$y(x) = (x + 1)(x + 1)$$  
    Expanding: $$y(x) = (x + 1)^2$$  
    **Derivative:**  
    $$y'(x) = 2(x + 1)$$  

11. $$y(x) = \frac{x}{x + 1}$$  
    Using the quotient rule:  
    $$y'(x) = \frac{(x+1)(1) - x(1)}{(x+1)^2} = \frac{1}{(x+1)^2}$$  

12. $$y(x) = (x + 1)\exp(x)$$  
    Using the product rule:  
    $$y'(x) = \exp(x) + (x+1)\exp(x) = (x+2)\exp(x)$$  

13. $$y(x) = \sin(x^2)$$  
    Using the chain rule:  
    $$y'(x) = \cos(x^2) \cdot 2x = 2x\cos(x^2)$$  

14. $$y(x) = \exp(-2x)$$  
    Using the chain rule:  
    $$y'(x) = -2\exp(-2x)$$  

15. $$y(x) = \frac{1}{\sin(x + 1)}$$  
    Using the chain rule and the derivative of $$\sin$$:  
    $$y'(x) = -\frac{\cos(x+1)}{\sin^2(x+1)}$$  

16. $$y(x) = 2x + 1$$  
    **Derivative:**  
    $$y'(x) = 2$$
